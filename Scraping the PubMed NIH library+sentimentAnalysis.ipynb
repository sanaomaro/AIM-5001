Now, start scraping for all the results we want, not only the first page of 10 resutls, through looping through pagination, the following function is for pagination process.


```python
#the paginatioon function to loop through pages
def func_main(year, keyword):
  #"https://pubmed.ncbi.nlm.nih.gov/?term=dangerous%5BTitle%5D&filter=dates.2021%2F1%2F1-2021%2F12%2F30&sort=date"
  uu = "https://pubmed.ncbi.nlm.nih.gov/?term="+keyword+"%5BTitle%5D&filter=dates."+year+"%2F1%2F1-"+year+"%2F12%2F30&sort=date"
  session = requests.Session()
  retry = Retry(connect=3, backoff_factor=0.5)
  adapter = HTTPAdapter(max_retries=retry)
  session.mount('http://', adapter)
  session.mount('https://', adapter)

  session.get(uu)
  browser.get(uu)
  pages_no = ()
  pages_no = browser.find_elements(By.ID, 'page-number-input')
  for i in pages_no:
    #print(i.get_attribute('max'))
    p = i.get_attribute('max')
   # page = page()
  pages = int(p)
  pages = np.arange(1,pages)
  return pages
```

**The next function scrape the results for all pages:**

```python
#the scraping function
def func(func_main,year, keyword):
  for page in func_main:
    uu = "https://pubmed.ncbi.nlm.nih.gov/?term="+keyword+"%5BTitle%5D&filter=dates."+year+"%2F1%2F1-"+year+"%2F12%2F30&sort=date&page="+str(page)
    session = requests.Session()
    retry = Retry(connect=3, backoff_factor=0.5)
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)

    session.get(uu)
    
    page = requests.get(uu)
    browser.get(uu)
   
    soup = BeautifulSoup(page.text, 'html.parser')
  
    box = soup.find('div', {'class': 'search-results-chunks'}) 
    list_links=browser.find_elements(By.CLASS_NAME, 'full-docsum')
    sleep(randint(2,8))
    for i in list_links:
      title = i.find_element(By.CLASS_NAME, 'docsum-title').text
      titles.append(title)
      #i.get_attribute("class")
      desc = i.find_element(By.CLASS_NAME, 'full-view-snippet').text
      descs.append(desc)
      id = i.find_element(By.CLASS_NAME, 'docsum-pmid').text
      ids.append(id)
      #  y = i.find_element(By.CLASS_NAME, 'docsum-journal-citation short-journal-citation').text #docsum-journal-citation full-journal-citation
      #//div[@class="docsum-citation full-citation"]//span[@class="docsum-journal-citation short-journal-citation"]//span[4]
      y = i.find_element(By.XPATH, '//div[@class="docsum-citation full-citation"]//span[4]').text
      years.append(y)
      urls = url_test.format(id)
      artic_links.append(urls)
      yearr.append(year)
      keywords.append(keyword)
```
