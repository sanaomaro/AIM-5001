First, we need to convert the data into long format, by making the three columns [Caltex	Gulf	Mobil] into one column [manufacturers_oil], and by putting their values as a seperated column called [No_gallons]

Second, we need to twist this new column No_gallons, by seperating it into two columns [purchased and consumed] because it does not make sense to have both values into one column.

**Steps followed to do that:**

1. Filling No_gallons values into a list.

```python
gallons=df['No_gallons'].tolist()
print(gallons)
print(type(gallons))
print(len(gallons))
```

2. Creating two lists, and using regex rules to seperate each value in the previous list and putting the value before the : into the the list that is going to fill the purchased column, and the number after ":" sign into the the list that is going to fill the consumed column.

```python
numbers = []
numbers1 = []
#gallons = df['No_gallons']
print(len(gallons))
print(gallons[0])
i = 0
for i in range(0,len(gallons)):
  numbers.append(re.findall('[0-9]+$',gallons[i]))
  numbers1.append(re.findall('^[0-9]+',gallons[i]))
 
type(numbers)
print(numbers)
print(numbers1)
```

```python
len(numbers1)
len(numbers)
```

3. removing any strings from these two newly created columns

```python
df['purchased'] = numbers1
```

```python
df['consumed'] = numbers
```

To get the required report, we need to divide the years into four periods, this could be done by a spcific function (get_year) and the n apply it to the group by function, after that applying the aggregation method to get the four columns of average, minimum, maximum, and count of LENGTH.

```python
def get_year(year):
  if year < 1860:
    return '1818-1860'
  elif year < 1902:
    return '1860-1902'
  elif year < 1944:
    return '1902-1944'
  else:
    return '1944-1986'

#df['Erected'] = df['ERECTED'].apply(get_year)
grouped = df.groupby(df['ERECTED'].apply(get_year))
#aggs = grouped.agg({'LENGTH': ['mean', 'count', 'min', 'max']})
aggs1 = grouped.agg(AverageLength=('LENGTH', 'mean'),
                   count=('LENGTH', 'count'),
                   Max_Length=('LENGTH', 'max'),
                   Min_Length=('LENGTH', 'min'))
#print(aggs)
print(aggs1)
```
